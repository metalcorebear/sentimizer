```python
# Instantiate SentiMizer
analyzer = SentiMizer()
```

```python
# Add text
analyzer.load_text(test_text)
```

```python
# Identify known entities
analyzer.find_entities()
```

```python
# list entities
print(analyzer.entities)
```

```python
# List sentences related to each entity
print(analyzer.sentences)
```

```python
# Measure emotional affect and sentiment
analyzer.emote()
```

```python
# Print emotions
print(analyzer.affect)
```

```python
# Print sentiment scores
print(analyzer.sentiments)
```

```python
# Plot entity sentiment scores

polarities = list(zip(analyzer.sentiments.keys(),  analyzer.sentiments.values()))
sorted_polarities = sorted(polarities, key=lambda i: i[1])

#get a color list
col_map = plt.get_cmap('tab10')
#plot figure
plt.figure(figsize=(12,8)) 
plt.title('Entity Sentiment Scores', fontsize=20 )
plt.barh(list(zip(*sorted_polarities))[0],list(zip(*sorted_polarities))[1], color = col_map.colors)
```

```python
# A simple script to plot top emotion scores and labels for each entity

entities_list = []
max_values = []
max_labels = []
for key in analyzer.affect:
    affect_dict = analyzer.affect[key]
    top_emotion = max(affect_dict, key=affect_dict.get)
    entities_list.append(key)
    max_values.append(top_emotion)
    max_labels.append(affect_dict[top_emotion])
    
#get a color list
col_map = plt.get_cmap('tab10')

#plot figure
plt.figure(figsize=(12,8)) 
plt.title('Entity Top Affect Scores', fontsize=20 )
plt.barh(entities_list, max_labels, color = col_map.colors)
for i, v in enumerate(max_labels):
    plt.text(v, i, max_values[i], va='center', color='blue')
```
